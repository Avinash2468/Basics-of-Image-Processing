```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import time
# import plotly
```

```python
# Custom built functions for image-IO operations

def read_image(path,a=1):
    if a==3:
        image = cv.imread(path)
        image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    if a==1:
        image = cv.imread(path, 0)
    return image

def show_image(img,text='',a=1):
    if a==3:
        plt.imshow(img)
    elif a==1:
        plt.imshow(img,cmap='gray')
    plt.axis("off")
    plt.title(text)
    plt.show()
    return 
    
def plot_figure(img_arr,a,b,text='',c=1):
    fig = plt.figure(figsize=(b*5,a*5+1))
    for i in range(1,a+1):
        for j in range(1,b+1):
            ax=fig.add_subplot(a,b,(i-1)*b+j)
            if c==3:
                ax.imshow(img_arr[(i-1)*b+j-1])
            if c==1:
                ax.imshow(img_arr[(i-1)*b+j-1],cmap='gray')
            ax.axis("off")
            ax.set_title("image - "+str((i-1)*b+j)+" size "+str(img_arr[(i-1)*b+j-1].shape), fontsize=12)
    
    plt.suptitle(text, fontsize=12)
    
    #ttl = plt.suptitle
    #ttl.set_position([.5, 1.05])
    #plt.tight_layout()
    #plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0,rect=[0, 0, 1, 0.])
    #fig.tight_layout(rect=[0, 0.01, 1, 0.95])
    plt.show()

    return 
```

## Q 3.1
Implement Bilateral Filtering and apply it on mountain.jpg

```python
"""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
FUNCTION    -    def filter_bilateral(img,k):

Inputs
img  -> required to be filtered
sigs -> bilateral edge preserving coefficient
sigc -> bilateral egde smoothing  coefficient

Outputs
out  -> filtered image 

No padding used, this is a filter that ouputs intensities only for valid pixel locations 
which have a complete overlap of neighbours with the filter kernel 

Works for only B/W images (1-channel images)

"""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""

def filter_bilateral(img, k,sigc=200,sigs=200):

    img = img.astype(float)
    l,b = img.shape
    sigc = -2*sigc*sigc ; sigs = -2*sigs*sigs
    
    out = np.zeros([l-k+1,b-k+1])
    coeff = np.zeros([l-k+1,b-k+1])
    
    for i in range(0,k):
        for j in range(0,k):

            f = np.exp( ((i-(k>>1))**2 + (j-(k>>1))**2)/sigc )
            temp = f * np.exp( ((img[(k>>1):l-(k>>1),(k>>1):b-(k>>1)] - img[i:i+l-k+1,j:j+b-k+1]))**2/sigs )
            
            coeff +=temp
            out += np.multiply( img[i:i+l-k+1,j:j+b-k+1] , temp )
    
    out = np.uint8(np.round(np.divide(out,coeff)))

    return out

"""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
FUNCTION    -    def color_bilateral(img,k):

Inputs
img  -> required to be filtered
sigs -> bilateral edge preserving coefficient
sigc -> bilateral egde smoothing  coefficient

Outputs
out  -> filtered image 

No padding used, this is a filter that ouputs intensities only for valid pixel locations 
which have a complete overlap of neighbours with the filter kernel 

Works for color images (3-Channel images)

"""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""


def color_bilateral(img,k,sigc=200,sigs=200):
    r,g,b = cv.split(img)

    fili_b = filter_bilateral(b,k, 100, 100)
    fili_g = filter_bilateral(g,k, 100, 100)
    fili_r = filter_bilateral(r,k, 100, 100)

    out = cv.merge((fili_r,fili_g,fili_b))
    
    return out

```
